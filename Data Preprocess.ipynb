```python
import pandas as pd

# 载入用户提供的 Dota 2 数据文件
file_path = 'dota2_matches.csv'
dota2_data = pd.read_csv(file_path)

# 显示数据的前几行以了解其结构
dota2_data.head()
```

```python
from datetime import datetime

# 移除可能不需要的列
columns_to_remove = ['version']
dota2_data_cleaned = dota2_data.drop(columns=columns_to_remove)

# 转换时间戳为易读的日期格式
dota2_data_cleaned['start_time'] = pd.to_datetime(dota2_data_cleaned['start_time'], unit='s')

# 计算 KDA（击杀-死亡-助攻）比率，注意避免除以零
dota2_data_cleaned['KDA'] = (dota2_data_cleaned['kills'] + dota2_data_cleaned['assists']) / dota2_data_cleaned['deaths'].replace(0, 1)

# 标记胜负
# 如果 player_slot 小于 128 且 radiant_win 为 True，或者 player_slot 大于等于 128 且 radiant_win 为 False，则为胜利
dota2_data_cleaned['win'] = ((dota2_data_cleaned['player_slot'] < 128) & dota2_data_cleaned['radiant_win']) | \
                            ((dota2_data_cleaned['player_slot'] >= 128) & ~dota2_data_cleaned['radiant_win'])

# 显示处理后的数据的前几行
dota2_data_cleaned.head()
```

```python
import requests

# Dota 2 API Endpoint to fetch hero information
url = "https://api.opendota.com/api/heroes"

# Sending a request to the API to get hero data
response = requests.get(url)

# Initialize an empty dictionary for hero ID to name mapping
hero_id_to_name = {}

if response.status_code == 200:
    heroes_data = response.json()
    # Create a mapping of hero ID to hero name
    hero_id_to_name = {hero['id']: hero['localized_name'] for hero in heroes_data}
else:
    print("Failed to fetch heroes data. Status Code:", response.status_code)

# Let's see a small part of the mapping
list(hero_id_to_name.items())[:5]  # Displaying first 5 mappings as an example

```

```python
# 应用映射将 hero_id 替换为英雄名称
dota2_data_cleaned['hero_name'] = dota2_data_cleaned['hero_id'].map(hero_id_to_name)

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 添加新列 'team' 来表示玩家所在的队伍
dota2_data_cleaned['team'] = dota2_data_cleaned['player_slot'].apply(lambda x: 'Radiant' if x < 128 else 'Dire')

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 映射表：将 game_mode 和 lobby_type 的数字代码转换为文字描述
game_mode_names = {
    1: 'All Pick', 2: 'Captain\'s Mode', 3: 'Random Draft', 4: 'Single Draft',
    5: 'All Random', 22: 'Ranked All Pick'
    # 这里只包括了一些常见的模式，可以根据需要添加更多
}

lobby_type_names = {
    0: 'Normal', 1: 'Practice', 2: 'Tournament', 3: 'Tutorial', 4: 'Co-op with bots',
    5: 'Team match', 6: 'Solo Queue', 7: 'Ranked'
    # 同样，这里只包括了一些常见的类型
}

# 应用映射将数字代码替换为文字描述
dota2_data_cleaned['game_mode_name'] = dota2_data_cleaned['game_mode'].map(game_mode_names)
dota2_data_cleaned['lobby_type_name'] = dota2_data_cleaned['lobby_type'].map(lobby_type_names)

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 映射表：将 game_mode 和 lobby_type 的数字代码转换为文字描述
game_mode_names = {
    1: 'All Pick', 2: 'Captain\'s Mode', 3: 'Random Draft', 4: 'Single Draft',
    5: 'All Random', 22: 'Ranked All Pick'
    # 这里只包括了一些常见的模式，可以根据需要添加更多
}

lobby_type_names = {
    0: 'Normal', 1: 'Practice', 2: 'Tournament', 3: 'Tutorial', 4: 'Co-op with bots',
    5: 'Team match', 6: 'Solo Queue', 7: 'Ranked'
    # 同样，这里只包括了一些常见的类型
}

# 应用映射将数字代码替换为文字描述
dota2_data_cleaned['game_mode_name'] = dota2_data_cleaned['game_mode'].map(game_mode_names)
dota2_data_cleaned['lobby_type_name'] = dota2_data_cleaned['lobby_type'].map(lobby_type_names)

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 删除指定的列：player_slot, game_mode, lobby_type, hero_id
columns_to_delete = ['player_slot', 'game_mode', 'hero_id','lobby_type']
dota2_data_cleaned = dota2_data_cleaned.drop(columns=columns_to_delete)

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 删除 'radiant_win' 列
dota2_data_cleaned = dota2_data_cleaned.drop(columns=['radiant_win'])

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 将 'duration' 列的格式从秒转换为 '分钟:秒'
dota2_data_cleaned['duration'] = dota2_data_cleaned['duration'].apply(lambda x: f"{x // 60}:{x % 60:02d}")

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 映射表：将 average_rank 的数字代码转换为 Dota 2 的具体排名名称
rank_names = {
    range(1, 11): 'Herald',
    range(11, 21): 'Guardian',
    range(21, 31): 'Crusader',
    range(31, 41): 'Archon',
    range(41, 51): 'Legend',
    range(51, 61): 'Ancient',
    range(61, 71): 'Divine',
    range(71, 81): 'Immortal'
}

# 应用映射将数字代码替换为排名名称
dota2_data_cleaned['average_rank_name'] = dota2_data_cleaned['average_rank'].apply(
    lambda x: next((name for ranks, name in rank_names.items() if x in ranks), 'Unknown'))

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 将 KDA 的数据保留到小数点后两位
dota2_data_cleaned['KDA'] = dota2_data_cleaned['KDA'].round(2)

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 转换 party_size 列为描述性的文本
dota2_data_cleaned['party_size_desc'] = dota2_data_cleaned['party_size'].apply(
    lambda x: 'Solo' if x == 1 else f'Party of {int(x)}' if x > 1 else 'Unknown'
)

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 删除 'party_size' 列
dota2_data_cleaned = dota2_data_cleaned.drop(columns=['party_size'])

# 解释 skill 和 leaver_status 的含义
# skill: 通常代表玩家的技能等级。可能包含的值及其含义可能有所不同，但通常是如下：
# - 1: Normal Skill
# - 2: High Skill
# - 3: Very High Skill

# leaver_status: 表示玩家在比赛结束时的状态。
# - 0: 正常完成比赛
# - 1: 离开了比赛，但比赛被标记为已完成
# - 2: 断开连接，但重新连接并完成了比赛
# - 3: 断开连接且没有重新连接
# - 4: 未知

# 转换 skill 和 leaver_status 列
skill_descriptions = {
    1: 'Normal Skill',
    2: 'High Skill',
    3: 'Very High Skill'
}

leaver_status_descriptions = {
    0: 'Finished',
    1: 'Left, but completed',
    2: 'Disconnected, but reconnected',
    3: 'Disconnected',
    4: 'Unknown'
}

dota2_data_cleaned['skill_desc'] = dota2_data_cleaned['skill'].map(skill_descriptions).fillna('Unknown')
dota2_data_cleaned['leaver_status_desc'] = dota2_data_cleaned['leaver_status'].map(leaver_status_descriptions)

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 删除 'skill' 和 'leaver_status' 列
dota2_data_cleaned = dota2_data_cleaned.drop(columns=['skill', 'leaver_status'])

# 显示处理后的数据的前几行来确认变化
dota2_data_cleaned.head()
```

```python
# 将处理后的 DataFrame 保存为 CSV 文件
csv_file_path = 'dota2_preprocessed.csv'
dota2_data_cleaned.to_csv(csv_file_path, index=False)

csv_file_path
```
